```python
import matplotlib.pyplot as plt
import numpy as np
from datetime import date
from random import random
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
from statsmodels.tsa.arima_model import ARMA
from statsmodels.graphics.api import qqplot

df = pd.read_csv(r"C:/Users/20966/Desktop/融合算法/trend - ARIMA.csv")
data1 = pd.read_csv(r"C:/Users/20966/Desktop/融合算法/trend - ARIMA.csv")
df = df['trend']
data = list(df)
df=list(df)
df=pd.Series(df)
data1=data1.set_index('year')
data1.index=pd.to_datetime(data1.index)
df.index = data1.index
df.plot()
plt.xlabel('year')
plt.ylabel('trend')
plt.show()
```

```python
date = pd.read_csv(r"C:/Users/20966/Desktop/date.csv")
date =date.set_index('year')
date_index=pd.to_datetime(date.index)

#指数平滑
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
# fit model
model = SimpleExpSmoothing(data)
model_fit = model.fit()
# make prediction
model_forecast_1 = model_fit.forecast(215)
print('用SES近220Day的预测值：',list(model_forecast_1))
model_forecast_1=pd.Series(model_forecast_1)
model_forecast_1.index = pd.Index(date_index)
model_forecast_1.plot(c="green", marker='o')
    
data1=data.copy()
data1=pd.Series(data1)
#data_index = sm.tsa.datetools.dates_from_range('1970','2021')
data1.index = pd.Index(date_index)
data1.plot()
    
plt.xlabel('year')
plt.ylabel('CO2')
plt.title('Exponential Smoothing')
plt.show()
```

```python
# Holt-Winters 法
# HWES
from statsmodels.tsa.holtwinters import ExponentialSmoothing
# contrived dataset
df = pd.read_csv(r"C:\Users\20966\Desktop\2023美赛的寒假集训命题（谢强军 2023.1.11）\Q3数据\Australia.CSV")
df = df['co2']
data = list(df)
# fit model
model = ExponentialSmoothing(data)
model_fit = model.fit()
#predict
model_forecast_1 = model_fit.forecast(10)
print('用HWES近十年的预测值：',list(model_forecast_1))
model_forecast_1=pd.Series(model_forecast_1)
data_index = sm.tsa.datetools.dates_from_range('2022','2031')
model_forecast_1.index = pd.Index(data_index)
model_forecast_1.plot(c="green", marker='o')
    
data1=data.copy()
data1=pd.Series(data1)
data_index = sm.tsa.datetools.dates_from_range('1970','2021')
data1.index = pd.Index(data_index)
data1.plot()
    
plt.xlabel('year')
plt.ylabel('CO2')
plt.title('HWES')
plt.show()
```

```python
# 自回归 (AR)
# AR
from statsmodels.tsa.ar_model import AutoReg
# contrived dataset
df = pd.read_csv(r"C:\Users\20966\Desktop\2023美赛的寒假集训命题（谢强军 2023.1.11）\Q3数据\Australia.CSV")
df = df['co2']
data = list(df)
# fit model
model = AutoReg(data, lags=1)
model_fit = model.fit()
# make prediction
model_forecast_1 = model_fit.forecast(10)
print('用AR近十年的预测值：',list(model_forecast_1))
model_forecast_1=pd.Series(model_forecast_1)
data_index = sm.tsa.datetools.dates_from_range('2022','2031')
model_forecast_1.index = pd.Index(data_index)
model_forecast_1.plot(c="green", marker='o')
    
data1=data.copy()
data1=pd.Series(data1)
data_index = sm.tsa.datetools.dates_from_range('1970','2021')
data1.index = pd.Index(data_index)
data1.plot()
    
plt.xlabel('year')
plt.ylabel('CO2')
plt.title('AR')
plt.show()
```

```python
# 季节性 ARIMA  (SARIMA)
# SARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
# contrived dataset
df = pd.read_csv(r"C:\Users\20966\Desktop\2023美赛的寒假集训命题（谢强军 2023.1.11）\Q3数据\Australia.CSV")
df = df['co2']
data = list(df)
# fit model
model = SARIMAX(data, order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
model_fit = model.fit(disp=False)
# make prediction
model_forecast_1 = model_fit.forecast(10)
print('用SARIMA近十年的预测值：',list(model_forecast_1))
model_forecast_1=pd.Series(model_forecast_1)
data_index = sm.tsa.datetools.dates_from_range('2022','2031')
model_forecast_1.index = pd.Index(data_index)
model_forecast_1.plot(c="green", marker='o')
    
data1=data.copy()
data1=pd.Series(data1)
data_index = sm.tsa.datetools.dates_from_range('1970','2021')
data1.index = pd.Index(data_index)
data1.plot()
    
plt.xlabel('year')
plt.ylabel('CO2')
plt.title('SARIMA')
plt.show()
```
